# Hangman game


```python
stages = ['''
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
=========
''', '''
  +---+
  |   |
  O   |
      |
      |
      |
=========
''', '''
  +---+
  |   |
      |
      |
      |
      |
=========
''']
```

```python
word_list = ['apple', 'banana', 'orange', 'grape', 'python', 'programming', 'cognizant', 'cricket', 'gadget']
```

```python
import random
logo = ''' 
 _                                             
| |                                            
| |__   __ _ _ __   __ _ _ __ ___   __ _ _ __  
| '_ \ / _` | '_ \ / _` | '_ ` _ \ / _` | '_ \ 
| | | | (_| | | | | (_| | | | | | | (_| | | | |
|_| |_|\__,_|_| |_|\__, |_| |_| |_|\__,_|_| |_|
                    __/ |                      
                   |___/    '''

print(logo)
life=len(stages)
word=list(random.choice(word_list))
# print(word)

result=[]
for i in range(len(word)):
    result.append('_')
print(result)
while('_' in result and life>0):
    choice=input("choose a word : ").lower()
    found=0
    for i in range(len(word)):
        if word[i]==choice:
            #print("correct")
            result[i]=choice
            found=1
       
    print(result)
    if found==0:
        life-=1
        print(stages[life])
    if life==0:
        print("game over")
        break
        
    
    
    
```
